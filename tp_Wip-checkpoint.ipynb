## Extracting txt and Word

```python
with open('test.txt') as f:
    input_text = f.read()

#input_text = open('test.txt', mode="r").read()

#Convert to same case -lower
lower_text = input_text.lower()
    
#tokenize
word_bag = nltk.tokenize.word_tokenize(lower_text)


#basic variables
unique_word = set(word_bag)
unique_wordCount = len(set(unique_word))
Total_wordCount = len(word_bag)
diversity_score = unique_wordCount / Total_wordCount

#Frequency distribution
fdist=FreqDist(word_bag)

print('The total word\punctuations usage =', Total_wordCount)
print('Number of unique words used =', unique_wordCount)
print('Word diversity score = ', diversity_score)


print(fdist)



# Read text

stopwords = STOPWORDS

wc = WordCloud(background_color="white", stopwords=stopwords, height=600, width=400)
wc.generate(lower_text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()

# store to file
#wc.to_file("wordcloud_output.png")
   
```

# Beautifulsoup and text extraction from website

```python
# Import packages
import requests
from bs4 import BeautifulSoup
###########################################################################################################
#Extracting text from a aURL
###########################################################################################################
# Specify url: url
url = 'https://ethereum.org/en/whitepaper/'

# Package the request, send the request and catch the response: r
r = requests.get(url)

# Extract the response as html: html_doc
html_doc = r.text

# Create a BeautifulSoup object from the HTML: soup
soup = BeautifulSoup(html_doc)

# Get the title of the webpage: guido_title
ETH_title = soup.title
# Print the title of Guido's webpage to the shell
print(ETH_title)
# Get Guido's text: guido_text

ETH_webCrawl = soup.get_text()
# Print Guido's text to the shell
eth_whitePaper = ETH_webCrawl[2545:85275]
###########################################################################################################
eth_whitePaper_LowerCase = eth_whitePaper.lower()
eth_token = nltk.tokenize.word_tokenize(eth_whitePaper_LowerCase)


len(set(eth_token))
print(len(eth_token))

print(type(eth_token))

#Generating a WOrd Cloud

stopwords = STOPWORDS

wc = WordCloud(background_color="white", stopwords=stopwords, height=600, width=400)
wc.generate(ETH_webCrawl)

# Display the generated image:
# the matplotlib way:
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
